### SSS - Twitter Data ###

*Super simple script (SSS) on getting data from Twitter*

Import the libraries

```python
import tweepy
```

Assign your own API keys in the constants below

```python
# Enter your own keys from your twitter account. 
API_KEY =''
API_SECRET =''
ACCESS_TOKEN =''
ACCESS_TOKEN_SECRET =''
```

Enter the keys above into tweepy's API

```python
auth = tweepy.OAuthHandler(API_KEY, API_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

api = tweepy.API(auth)
```

List of users which you would like to get feeds from

```python
users = ['scottadamssays']
```

Get tweets from the user's timeline

```python
user_timeline = api.user_timeline(screen_name=users[0], count=10)
```

```python
for idx, tweet in enumerate(user_timeline):
    print(user_timeline[idx].text)
```

Hashtags that you would like to search

```python
topics = ['#infinitywars']
```

Use the search API

```python
for tweet in tweepy.Cursor(api.search, 
                           q=topics[0], 
                           count=100,
                           lang='en',
                           since='2018-01-01').items():
    print(tweet.created_at, tweet.text)
```
