### SSS - Twitter Sentiment ###

*Super simple script (SSS) on sentiment scoring of data from Twitter*

Import the libraries

```python
import tweepy
from textblob import TextBlob
```

Assign your own API keys in the constants below

```python
# Enter your own keys from your twitter account. 
API_KEY =''
API_SECRET =''
ACCESS_TOKEN =''
ACCESS_TOKEN_SECRET =''
```

Enter the keys above into tweepy's API

```python
auth = tweepy.OAuthHandler(API_KEY, API_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

api = tweepy.API(auth)
```

List of users which you would like to get feeds from

Hashtags that you would like to search

```python
topics = ['#tradewar']
```

Use the search API

```python
for tweet in tweepy.Cursor(api.search, 
                           q=topics[0], 
                           count=100,
                           lang='en',
                           since='2018-08-10').items():
    tweetblob = TextBlob(tweet.text)
    print(tweet.created_at)
    print(tweet.text)
    print(tweetblob.sentiment.polarity)
```
